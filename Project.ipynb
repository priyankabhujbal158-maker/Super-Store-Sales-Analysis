  ##  **Which are the top 5 product categories based on total sales amount?**


Answer-The code groups data by Product_Category and sums up the Amount for each category.A Seaborn barplot is created to visualize the top categories

```python
D1 = data.groupby("Product_Category",as_index=False)["Amount"].count().sort_values(by="Amount",ascending=False).head(5)

plt.figure(figsize=(15,6))

E1=['#2ecc71', '#58d68d', '#82e0aa', '#abebc6','#d5f5e3' ]

C1 = sns.barplot(x="Product_Category" , y="Amount", data=D1,palette=E1,edgecolor="black",linewidth=1.5)

plt.title("Product_category wise Amount",family="Georgia",fontsize=16,fontweight="bold",color="#255E3D")

plt.xlabel("")

plt.ylabel("")

for i in C1.containers:
    C1.bar_label(i)
plt.show()
```

  ##  **what is the Age Group wise Gender Distribution ?**

Answer-This chart displays the gender distribution across different age groups.it shows how many males and females are present in each group.


```python
plt.figure(figsize=(10,5))

D2=["#721c97", '#a569bd']

C3=sns.countplot(x="Age Group",data=data,hue="Gender",palette=D2)

plt.title("Age_Group wise Gender Distribution",family="Georgia",fontsize=16,fontweight="bold",color="black")

plt.xlabel("")

plt.ylabel("")

for i in C3.containers:
    C3.bar_label(i)
plt.show()
```

  ##  **what is the Gender count across different Zones ?**

Answer-This chart shows the number of males and females in each zone (Western,South,Central ,Northern,and Eastern)

```python
plt.figure(figsize=(8,5))

D1=['#e74c3c', '#ee796d', '#f2a29b', '#f6cfc6', '#fbeae8']

C2=sns.countplot(x="Gender",data=data,hue="Zone",palette=D1,edgecolor="black",linewidth=1)

plt.title("Gender wise Count",family="Georgia",fontsize=16,fontweight="bold",color="black")


for i in C2.containers:
    C2.bar_label(i)
```

 ##  **What is the Sector wise Orders Distribution ?**

Answer-This line chart shows how the number of orders varies by sectors.Each point on the line represents the total orders placed by people of that sectors.

```python
plt.figure(figsize=(15,6))

D1=["#9223c1"]

D5=[ '#a569bd']



sns.lineplot(x="Sector",y="Orders",data=data,errorbar=None,marker="o",color="#9223c1")

plt.title("Sector wise Orders",family="Georgia",fontsize=16,fontweight="bold",color="black") 

plt.xlabel("")

plt.xticks(rotation=30)

plt.ylabel("")

plt.grid(color='#a569bd')

plt.show()


```

  ##  **What is the Zone wise Amount Distribution ?**

Answer-A pie chart is used to show how much each zone contributes to the total sales.The autopct option shows percentage valueson the chart for clarity.

```python
D4=data.groupby(["Zone"],as_index=False)["Amount"].sum().sort_values(by="Amount",ascending=False)

print(D4)

E1=['#e67e22', '#eb984e', '#f0b27a', '#f5cba7','#fbe9e7']

plt.pie(D4["Amount"],labels=D4["Zone"],startangle=110,autopct="%1.f%%",colors=E1)

plt.title("Zone wise Amount",family="Georgia",fontsize=16,fontweight="bold",color="#ad580d")

plt.show()
```
